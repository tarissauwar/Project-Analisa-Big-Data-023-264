### Distribusi Data pada Kolom Kategorikal

```python
import pandas as pd
import matplotlib.pyplot as plt

print("\nDistribusi Data pada Kolom Kategorikal:\n")
for col in data.select_dtypes(include=['object']).columns:
    print(f"\nKolom: {col}")
    value_counts = data[col].value_counts()

    # Tampilkan distribusi dalam bentuk tabel
    print(value_counts)

    # Visualisasi bar chart
    plt.figure(figsize=(10, 6))
    value_counts.plot(kind='bar', color='skyblue', edgecolor='black')
    plt.title(f"Distribusi Data untuk Kolom: {col}")
    plt.xlabel(col)
    plt.ylabel("Jumlah")
    plt.xticks(rotation=45)
    plt.tight_layout()
    plt.show()
```

## Handling Missing Value

```python
# 1. Isi nilai yang hilang 'children' dengan median
data['children'].fillna(data['children'].median(), inplace=True)

# 2. Isi nilai yang hilang di 'country' dengan 'Unknown'
data['country'].fillna('Unknown', inplace=True)

# 3. Isi nilai yang hilang di 'agent' dengan 0 (dengan asumsi nilai yang hilang menunjukkan tidak ada agen)
data['agent'].fillna(0, inplace=True)

# 4. Fill missing values in 'company' with 0 (assuming missing values indicate no company)
data['company'].fillna(0, inplace=True)

# Check if there are any remaining missing values
missing_values_after = data.isnull().sum()
missing_values_after
```

## Cek Korelasi antar kolom numeri

```python
# Pilih hanya kolom numerik
numerical_columns = data.select_dtypes(include=['float64', 'int64']).columns

# Menghitung korelasi antar kolom numerik
corr_matrix = data[numerical_columns].corr()

# Membuat heatmap untuk visualisasi korelasi
plt.figure(figsize=(12, 8))
sns.heatmap(corr_matrix, annot=True, cmap='coolwarm', linewidths=0.5)
plt.title('Korelasi Antar Variabel Numerik')
plt.show()
```

```python
# Menentukan fitur yang memiliki korelasi tinggi (misalnya di atas 0.3 atau kurang dari -0.3)
high_corr_features = corr_matrix.columns[(corr_matrix.abs() > 0.3).any()]
print("Fitur dengan korelasi tinggi: ", high_corr_features)
```

## Analisis Faktor yang Mempengaruhi Tipe Hotel yang Dipilih oleh Pelanggan

```python
# Visualisasi dengan count plot untuk melihat distribusi
plt.figure(figsize=(10, 6))
ax = sns.countplot(data=data, x='arrival_date_year', hue='hotel', palette='Set2')

# Memberikan judul dan label
plt.title('Distribusi Tipe Hotel yang Paling Banyak Dikunjungi per Tahun', fontsize=16)
plt.xlabel('Tahun', fontsize=12)
plt.ylabel('Jumlah Reservasi', fontsize=12)

# Menambahkan angka di atas setiap bar
for p in ax.patches:
    ax.annotate(f'{int(p.get_height())}',
                (p.get_x() + p.get_width() / 2., p.get_height()),
                ha='center', va='center',
                fontsize=12, color='black',
                xytext=(0, 5), textcoords='offset points')

# Menampilkan grafik
plt.xticks(rotation=45)
plt.show()
```

## Analisa ketertarikan tipe pelanggan dalam memilih tipe hotel

## Analisa keberadaan tamu hotel yang datang kembali dalam pemilihan tipe hotel

```python
# Menghitung jumlah tamu berulang untuk setiap tipe hotel
repeated_guest_counts = data.groupby(['hotel', 'is_repeated_guest']).size().reset_index(name='counts')

# Mengubah nilai numerik di 'is_repeated_guest' menjadi kategori
repeated_guest_counts['is_repeated_guest'] = repeated_guest_counts['is_repeated_guest'].replace({0: 'Not Repeated', 1: 'Repeated'})

# Visualisasi data
plt.figure(figsize=(10, 6))
sns.barplot(data=repeated_guest_counts, x='hotel', y='counts', hue='is_repeated_guest', palette='coolwarm')

plt.title('Jumlah Tamu Berulang dan Tidak Berulang Berdasarkan Tipe Hotel', fontsize=14)
plt.xlabel('Tipe Hotel', fontsize=12)
plt.ylabel('Jumlah', fontsize=12)
plt.legend(title='Tipe Tamu')
plt.xticks(rotation=0)
plt.tight_layout()
plt.show()

# Analisis deskriptif
hotel_repeated_ratio = data.groupby('hotel')['is_repeated_guest'].mean().reset_index()
hotel_repeated_ratio['is_repeated_guest'] = hotel_repeated_ratio['is_repeated_guest'] * 100
print("Rasio Tamu Berulang per Tipe Hotel:")
display(hotel_repeated_ratio)
```

### Interpretasi Hasil


Mari amati bersama hasil dari analisa diatas pada tipe pelanggan di tiap tipe hotel.
1. City Hotel memiliki rasio tamu berulang sebesar **3.20%**
  *   Artinya, dari seluruh tamu yang menginap di City Hotel, hanya sekitar 3.20% adalah tamu yang pernah menginap sebelumnya (tamu berulang). Ini menunjukkan bahwa mayoritas tamu City Hotel adalah tamu baru.
1. Resort Hotel memiliki rasio tamu berulang sebesar **5.03%**
  *   Artinya, dari seluruh tamu yang menginap di Resort Hotel, sekitar 5.03% adalah tamu berulang. Meskipun lebih tinggi dibandingkan City Hotel, angka ini tetap menunjukkan bahwa sebagian besar tamu Resort Hotel juga merupakan tamu baru.

Berdasarkan hasil tersebut, terdapat beberapa pola yang dapat diamati
1. Resort Hotel cenderung memiliki tingkat loyalitas pelanggan yang sedikit lebih tinggi dibandingkan dengan City Hotel. Hal ini dapat disebabkan oleh beberapa faktor seperti layanan, pengalaman unik, atau lokasi yang lebih menarik bagi tamu untuk kembali.
2. City Hotel mungkin lebih sering digunakan untuk kunjungan singkat seperti perjalanan bisnis, di mana pelanggan lebih jarang kembali ke hotel yang sama karena sifat perjalanan yang lebih fleksibel.

## Pemilihan tipe hotel berdasarkan "Kid-Friendly atau Family-Oriented"

```python
# Menambahkan kolom untuk mengidentifikasi apakah reservasi dilakukan oleh keluarga
data['is_family'] = data[['children', 'babies']].sum(axis=1).apply(lambda x: 'Family' if x > 0 else 'Not Family')

# Menghitung jumlah reservasi keluarga dan non-keluarga untuk setiap tipe hotel
family_counts = data.groupby(['hotel', 'is_family']).size().reset_index(name='counts')

# Visualisasi data menggunakan pie chart
plt.figure(figsize=(14, 7))
colors = ['#E195AB', '#81BFDA']
explode = (0.1, 0)  # Efek explode untuk kategori Family

# Membuat subplot untuk setiap tipe hotel
hotels = data['hotel'].unique()
for i, hotel in enumerate(hotels):
    plt.subplot(1, len(hotels), i + 1)
    hotel_data = family_counts[family_counts['hotel'] == hotel]
    plt.pie(
        hotel_data['counts'],
        labels=hotel_data['is_family'],
        autopct='%1.1f%%',
        colors=colors,
        startangle=140,
        explode=explode,
        textprops={'fontsize': 12}
    )
    plt.title(f"{hotel}", fontsize=16, fontweight='bold')

# Menambahkan judul utama
plt.suptitle(
    'Distribusi Reservasi Keluarga dan Non-Keluarga Berdasarkan Tipe Hotel',
    fontsize=18,
    fontweight='bold',
    color='#333333'
)

plt.tight_layout(rect=[0, 0, 1, 0.95])
plt.show()
```

### Interpretasi Hasil

Mari amati hasil dari analisa diatas:
1. City Hotel
  *   Family: 9.8% dari total keseluruhan reservasi
  *   Not Family: 90.2% dari total keseluruhan reservasi
2. Resort Hotel
  *   Family: 11.4% dari total keseluruhan reservasi
  *   Not Family: 88.6% dari total keseluruhan reservasi

Berdasarkan hasil tersebut, dapat diambil kesimpulan bahwa **City Hotel** cenderung lebih diminati oleh tamu individu atau pasangan, memungkinkan preferensi pemilihan tipe hotel tersebut untuk perjalanan bisnis atau kunjungan singkat. Sedangkan, **Resort Hotel** persentase peminat **Family** lebih besar dibandingkan **City Hotel**. Artinya, **Resort Hotel** lebih diminati oleh keluarga, meskipun dalam proporsi yang cukup kecil.
